# Задача 1
### Условие
Далее приведен график объемов карточных трат клиентов за некоторый период
2022 года.
Предположи, что это могли быть за траты и почему они упали до 0?

### Решение
Есть два варианта - либо данные некорректны, либо корректны.

Если данные некорректны, возможно произошел сбой и в хранилище перестала загружаться часть данных. В таком слуяае, нужно проверить, что данные в хранилище соответствуют источнику.
Еще может быть, что источник стал по-новому размечать траты клиентов. Если в графике есть фильтр по типам трат, то изменившийся тип перестал отображаться на графике, то есть отсеялся фильтром. 

Если данные корректны, значит либо все пользователи резко поменяли свое поведение, либо случилось что-то, из-за чего они больше не могут совершать траты определенного типа. Врядли в большом банке 100% пользователей в один день вдруг полностью перестали совершать определенные траты. При этом по графику видно, что перед тем, как траты упали в ноль, они резко выросли. Вероятно, было введено какое-то ограничение на определенные траты, пользователи нарастили эти траты перед вводом ограничения, а потом перестали их совершать. Если погуглить новости начала марта 2022 можно найти, что 10 марта 2022 платежные системы Visa и Master Card прекратили обслуживание российских карт, и они перестали работать за рубежом.

### Ответ
На графике приведена статистика трат по российским картам Visa и Master Card за рубежом. Причина резкого роста и последующего падения трат до 0 это прекращение обслуживания карт платежными системами.
