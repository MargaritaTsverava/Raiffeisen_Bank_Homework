### Условие
Есть бизнес-задача подготовить клиентский оффер с повышенным кэшбэком на
категории, направленный на рост метрики Merch4.
Merch4 - % клиентов от активной базы, совершающих покупки по дебетовой
карте в 4+ разных категориях (пример мерчантов: аптеки, кафе,
супермаркеты и т.д)
Опиши, какие данные и их разрезы потребуются для анализа, чтобы определить
условия оффера, дающие максимальный эффект на Merch4.
Для анализа тебе доступны:

● данные по карточным операциям:
- дата операции
- сумма операции
- МСС-код
- Merchant name
- ID клиента
- Product name

● Данные по клиентам:
- ID клиента
- сегмент клиента
- флаг (новый/действующий клиент


### Решение
В задаче нет ограничений на механику предложения кэшбека.
В таком случае, мы, как и некоторые другие банки (альфа банк, тинькофф), можем предлагать каждому клиенту категории кэшбека на выбор. Например, каждый месяц клиент выбирает 3-5 категорий повышенного кэшбека.
Каждый раз перед тем как предложить клиенту категории кэшбека на выбор мы можем сделать следующие расчеты:
- Найти наиболее популярные категории кэшбека среди активных клиентов. Для этого найдем количество уникальных клиентов, которые совершили хотя бы одну покупку по каждой категории за прошедший период и отсортируем получившийся список по убыванию количества уникальных клиентов.
- Перед тем, как сделать клиенту предложение категорий кэшбека на выбор, найдем по нему категории, в которых он уже совершал траты в предыдущий период.
- Предложим клиенту самые популярные категории из первого списка за вычетом категорий, в которых он уже совершает покупки (второй список).

Опционально мы можем поработать над уменьшением оттока клиентов из категории Merch4. Для этого время от времени мы можем помимо категорий где у клиента трат вообще нет предлагать клиенту на выбор те категории, в которых он либо недавно перестал совершать покупки, либо стал покупать меньше по сравнению с предыдущими периодами. Иногда можно для мотивации клиента также предлагать ему на выбор категории, в которых он уже активно совершает траты.
